Given data on US flights from 2013 about the arrival and departure times and other info we want to create a model to give a percentage chance that a flight on a given day of the week and airport

```python
import pandas as pd

# Load the CSV file (adjust the path if needed)
df = pd.read_csv('./data/flights.csv')

# Show the first 5 rows
df.head()
```

```python
# ...existing code...

# Check for missing values in each column
print(df.isnull().sum())

# Get data types and a summary
print(df.info())
print(df.describe(include='all'))
# ...existing code...
```

```python
# ...existing code...

# Replace all missing values in the DataFrame with 0
df = df.fillna(0)

# Check again to confirm there are no missing values
print(df.isnull().sum())
# ...existing code...
```

```python
# ...existing code...

# Select features and target
# We'll use DayOfWeek and OriginAirportID as features, and ArrDel15 as the target
features = df[['DayOfWeek', 'OriginAirportID']]
target = df['ArrDel15']

# Show the first few rows of features and target
print(features.head())
print(target.head())
# ...existing code...
```

```python
# ...existing code...

from sklearn.model_selection import train_test_split

# Split the data: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)

# Show the shape of the splits
print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
# ...existing code...
```

```python
# ...existing code...

from sklearn.linear_model import LogisticRegression

# Create and train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Show training complete
print("Model training complete.")
# ...existing code...
```

```python
# ...existing code...

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Predict on the test set
y_pred = model.predict(X_test)

# Print accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Print classification report (precision, recall, f1-score)
print("Classification Report:\n", classification_report(y_test, y_pred))

# Print confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
# ...existing code...
```

```python
# ...existing code...

# Predict the probability of delay (ArrDel15 = 1) for the test set
y_proba = model.predict_proba(X_test)[:, 1]

# Show the first 10 predicted probabilities
print("Predicted probability of delay for first 10 test samples:")
print(y_proba[:10])
# ...existing code...
```

```python
# ...existing code...

# Example: Predict probability for Monday (1) at Tampa International (15304)
import numpy as np
input_data = pd.DataFrame({
    'DayOfWeek': [1],
    'OriginAirportID': [15304]
})

# Predict probability of delay (ArrDel15 = 1)
probability = model.predict_proba(input_data)[:, 1][0]
print(f"Predicted probability of delay: {probability:.2%}")
# ...existing code...
```

```python
# ...existing code...

import joblib

# Save the trained model to a file
joblib.dump(model, 'flight_delay_model.joblib')
print("Model exported as flight_delay_model.joblib")
# ...existing code...
```

```python
import pandas as pd

# Load the flights data
df = pd.read_csv('./data/flights.csv')

# Get unique Origin airports
origin_airports = df[['OriginAirportID', 'OriginAirportName', 'OriginCity', 'OriginState']].drop_duplicates()
origin_airports = origin_airports.rename(columns={
    'OriginAirportID': 'AirportID',
    'OriginAirportName': 'AirportName',
    'OriginCity': 'City',
    'OriginState': 'State'
})

# Get unique Dest airports
dest_airports = df[['DestAirportID', 'DestAirportName', 'DestCity', 'DestState']].drop_duplicates()
dest_airports = dest_airports.rename(columns={
    'DestAirportID': 'AirportID',
    'DestAirportName': 'AirportName',
    'DestCity': 'City',
    'DestState': 'State'
})

# Combine and drop duplicates
all_airports = pd.concat([origin_airports, dest_airports]).drop_duplicates().sort_values('AirportID')

# Save to CSV
all_airports.to_csv('./data/airports.csv', index=False)
print("Exported airports to ./data/airports.csv")
```
